```python
from bs4 import BeautifulSoup
from selenium import webdriver
```

```python
def get_elements(url, start, end):
    driver = webdriver.Chrome()
    driver.get(url)
    
    #from hidden elements
    innertab = driver.find_element_by_xpath('//*[@id="container"]/div[4]/div[2]/div[3]')
    hidden = innertab.get_property('innerHTML')
    dom = BeautifulSoup(hidden, 'html.parser')
    
    #lecture title
    lec_title = []
    for _ in dom.find_all('strong')[start+3:end+4]:#,{'class':'sub_new'})
        lec_title.append(_.text)
        
    #descrip $ keyword
    descrip = []
    keyword = []
    for _ in dom.find_all('div',{'class':'con_wrap'})[start:end+1]:
        for i in [_.find_all('dd')]:
            descrip.append(i[0].text)
            keyword.append(i[1].text)
            
    #image 
    img = []
    course_id = []
    lecture_id = []
    for _ in dom.find_all('p',{'class':'sub_thumb'})[start:end+1]:
        for i in _.find_all('img'):
            img.append(i['src'])
            
    #video & courseid & lecture_id
    video = []
    course_id = []
    lec_id = []
    dom2 = driver.find_element_by_xpath('//*[@id="preList"]')
    for _ in list(reversed(dom2.find_elements_by_tag_name('option')))[start:end+1]:
        video.append(_.get_attribute('value'))
        course_id.append(_.get_attribute('data-courseId'))
        lec_id.append(_.get_attribute('data-lectid'))
            
    return lec_id, course_id, lec_title, img, video, descrip, keyword  

           

    
    
    
```

```python
def get_result(start, end):
    filenames = []
    text = ''
    for i in range(start, end+1):
        n = '{0:05d}'.format(i)
        result_fname = 'result_conversion_' + n +'.txt'
        filenames.append(result_fname)
        for files in filenames:
            with open(files,'r', encoding = 'euc-kr') as r:
                text+=r.read()
            
    return text
```
