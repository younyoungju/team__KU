```python
def get_result(start, end):
    filenames = []
    text = ''
    for i in range(start, end+1):
        n = '{0:05d}'.format(i)
        result_fname = 'result_conversion_' + n +'.txt'
        filenames.append(result_fname)
        for files in filenames:
            with open(files,'r', encoding = 'euc-kr') as r:
                text += r.read()
            
    return text.split('\n')[1:]
```

```python
all_text = []
for i in range(1,26):
    all_text.append(get_result(i,i+1))
```

```python
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect('LectList.db')
cur = conn.cursor()
```

```python
for i in all_text:
    cur.executescript('''
        ALTER TABLE Lec_list ADD COLUMN Result;
        INSERT INTO Lec_list (Result) values(i);
    ''')
```

```python
cur.execute(''' SELECT * FROM Lec_list''')
```
