```python
import json
```

```python
def getText_json(file):
    with open(file, encoding='UTF-8-sig') as f:
        data = json.load(f)
    for _ in data['results']['transcripts']:
         para = "".join(_['transcript'])
    return para
```

```python
compare = getText_json('asrOutput(1).json')
compare
```

```python
import re
```

```python
def getText_smi( File ):
    text = []
    file = open(File, "r" , encoding = "euc_kr")
    wholefile = file.readlines()
    for line in wholefile:
        line = line.strip()
        if re.search(r"#",line) or line.endswith(":>"):
            continue
        elif line.startswith('<') or line.endswith("; }"):
            continue
        elif line.startswith('—') or line.endswith(";"):
            continue
        text.append(line)   #re 쓰는법 알려주세요
    return ''.join(text)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# 객체 생성
tfidf_vectorizer = TfidfVectorizer()

# 문장 벡터화 진행
tfidf_matrix = tfidf_vectorizer.fit_transform(sentence)

# 각 단어
text = tfidf_vectorizer.get_feature_names()

# 각 단어의 벡터 값
idf = tfidf_vectorizer.idf_
```
