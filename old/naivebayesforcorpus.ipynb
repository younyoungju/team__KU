```python
import re

def EBS_dict(s, e):
    pattern1 = re.compile(r'\s{2,}') 
    pattern2 = re.compile(r'[가-힣.?,‘’]') 
#     pattern3 = re.compile(r'[ㄱ-ㅎa-z]{1}') # ㄱ,ㄴ ....a,b,c....
    
    dic = []
    for i in range(s, e+1):
        n = '{0:05d}'.format(i)
        lec = 'lec' + n + '_EBS.txt'
        dic.append(lec)
        
    texts = ''
        
    for _ in dic:
        with open('data/'+ _ ,'r',encoding='utf-8') as r:
            texts += r.read()

    texts = pattern1.sub(' ', texts) ## 
    texts = pattern2.sub(' ', texts) ## 
#     texts = pattern3.sub(' ', texts) ## 
#     EBSDic = texts.split()
    EBSDic = texts
    return list(set(EBSDic.split()))
#     return ' '.join(EBSDic.split())
    
#     return [_ for _ in set(EBSDic) if len(_) == 1] + ['sin','cos','tan']
```

```python
def google_text(s,e):
    
    dic = []
    for i in range(s, e+1):
        n = '{0:05}'.format(i)
        lec = 'lec' + n + '_STT1_Google.txt'
        dic.append(lec)
        
    texts = ''
    
    for _ in dic:
        with open('data/'+ _ , 'r', encoding = 'euc-kr') as r:
            texts += r.read()
    return texts
```

```python
def google_word_text(s,e):
    
    dic = []
    for i in range(s, e+1):
        n = '{0:05}'.format(i)
        lec = 'lec' + n + '_STT1_Google.txt'
        dic.append(lec)
        
    import re    
    wordtoken = []
    
    for _ in dic:
        with open('data/'+ _ , 'r', encoding = 'euc-kr') as f:
            for line in f:
                for word in re.findall(r'\w+', line):
                    wordtoken.append(word)
    return wordtoken
```
