```python
def selected_pos(pos_text):
    alist = []
    for _ in pos_text:
        if _[1] in ['NNG','NNP','NNB','NNM', 'NR','NNA', 'NP', 
                    'VV','VA', 'VX', 'VXV', 'VXA', 'VCP', 'VCN', 'VX',
                    'MDT','MDN', 'MAG', 'MAC', 'MD',  
                    'JKS', 'JKC', 'JKG', 'JKO', 'JKM', 'JKI', 'JKQ', 'JX', 'JC', 
                    'IC', 
                    'OL',
                    'IC',
                    'XR']:
            alist.append(_[0])
            continue
    return alist
```

```python
list_EBS_pos = selected_pos(ebs_pos)
list_google_pos = selected_pos(google_pos)
```
