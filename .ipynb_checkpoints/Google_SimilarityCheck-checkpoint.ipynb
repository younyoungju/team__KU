```python
def google_text(s,e):
    
    dic = []
    for i in range(s, e+1):
        n = '{0:05}'.format(i)
        lec = 'lec'+n+'_STT1_Google.txt'
        dic.append(lec)
        
    texts = ''
    for _ in dic:
        with open('.\\STT1_Google\\' + _ , 'r', encoding = 'euc-kr') as r:
            texts += r.read()
    return texts
```

```python
textG = google_text(1,26)   #26개의 구글 stt
```

```python
#EBS = okt.nouns(textE)
Google = okt.nouns(textG)
```

```python
from collections import Counter
noun_counter = Counter([noun for text in Google for noun in okt.nouns(text)]
)

print('num of nouns: %d' % len(noun_counter))
```

```python
for min_count in [100,150,200,250]:   #30,50,70,100
    _counter = {
        word for word, freq in noun_counter.items()
        if freq >= min_count
    }
    print('num of nouns (min_count = %d): %d %s' % (
        min_count, len(_counter), _counter))
```

```python
#30번 이상 나오는 266 개의 명사
#sin = 싸인, 사인, 
```

```python
noun_dict = {
    word for word, freq in noun_counter.items()
    if freq >= 70
}

def custom_tokenizer(doc):
    return [word for word in okt.nouns(doc) if word in noun_dict]

print(okt.nouns(Google[1]))
print(custom_tokenizer(Google[1]))
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(
    tokenizer=custom_tokenizer,
    min_df=0.005,
    max_df=0.95
)

x_sparse = vectorizer.fit_transform(Google)
```
