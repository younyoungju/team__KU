```python
def read_file(s, e, d): #Amazon Transcribe STT파일 불러오는 코드
    lecs = []
    data = ''
    for i in range(s, e+1):
        n = '{0:05d}'.format(i)
        lec = 'lec' + n + '_STT1_Transcribe.txt'
        lecs.append(lec)
    for _ in lecs:
        file_directory = d + _
        with open(file_directory, 'r', encoding='utf-8') as f:
            print(f.read())
            data += f.read()
    return data
```
